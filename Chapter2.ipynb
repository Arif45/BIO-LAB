```python
import itertools
def substrings(k):
    sub = map(''.join, itertools.product('ACTG', repeat=k))
    answers = dict()
    for i in sub:
        answers[i] = 0

    return sorted(answers)

def hamming_dist(str1, str2):

    l1 = len(str1)
    l2 = len(str2)

    c = 0
    for i in range(l1):
        if str1[i] != str2[i]:
            c += 1

    return c

def min_distance(dna, pattern):
    a = list()
    start = 0
    length = len(dna)
    k = len(pattern)
    while start + k <= length:
        str = dna[start:start+k]
        a.append(hamming_dist(str, pattern))
        start += 1

    return min(a)

def distance_in_str(dna, pattern):
    distances = list()
    for x in dna:
        distances.append(min_distance(x, pattern))

    return sum(distances)

def find_median_string(dna, k):
    patterns = substrings(k)
    answers = dict()
    for k in patterns:
        answers[k] = distance_in_str(dna, k)

    print( min(answers, key=answers.get))

dna = ["AAATTGACGCAT",
"GACGACCACGTT",
"CGTCAGCGCCTG",
"GCTGAGCACCGG",
       "TACGGGACAG"]
k=3
find_median_string(dna,3)
```
